```python
import re
import pytesseract
from pytesseract import Output
import cv2
from matplotlib import pyplot as plt
import os
```

```python
os.chdir(r'C:\Users\Games king\Desktop\OCR')
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

def extract_text_and_data_upper(image_path):
 img = cv2.imread(image_path)
 gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
 _, threshold_img = cv2.threshold(gray_img, 150, 255, cv2.THRESH_BINARY)
 
 custom_config = r'--oem 3 --psm 6'
 results = pytesseract.image_to_string(threshold_img, config=custom_config)
 upper_results = results.upper()
 print("النص المستخرج بالحروف الكبيرة:")
 print(upper_results)

 name_match = re.search(r'NAME:s*(.*)', upper_results)
 name = name_match.group(1).strip() if name_match else 'اسم غير موجود'

 
 id_match = re.search(r'ID:s*(S+)', upper_results)
 id_number = id_match.group(1).strip() if id_match else 'ID غير موجود'

 
 plt.imshow(threshold_img, cmap='gray')
 plt.title('Processed Image')
 plt.show()

 
 return name, id_number

image_path = 'photo_2024-04-09_09-44-45.jpg'
extracted_name, extracted_id = extract_text_and_data_upper(image_path)
print("الاسم المستخرج :", extracted_name)
print("الرقم ا المستخرج :", extracted_id)
```
