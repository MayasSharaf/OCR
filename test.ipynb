```python
import pytesseract
from PIL import Image, ImageEnhance, ImageFilter, ImageOps
import os
import cv2
import numpy as np

os.chdir(r'C:\Users\Games king\Desktop\OCR')
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

def ocr_image_to_text(image_path, lang='ara'):
    try:
        # Open an image file
        img = Image.open(image_path)

        # Convert the image to grayscale
        gray_img = img.convert('L')

        # Apply Gaussian blur to reduce noise
        gray_img = np.array(gray_img)
        gray_img = cv2.GaussianBlur(gray_img, (5, 5), 0)

        # Apply thresholding to binarize the image
        _, binary_img = cv2.threshold(gray_img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

        # Apply dilation and erosion to remove noise and fill in gaps
        kernel = np.ones((3, 3), np.uint8)
        binary_img = cv2.dilate(binary_img, kernel, iterations=1)
        binary_img = cv2.erode(binary_img, kernel, iterations=1)

        # Convert the binary image back to grayscale
        gray_img = cv2.cvtColor(binary_img, cv2.COLOR_BGR2GRAY)
        gray_img = Image.fromarray(gray_img)

        # Invert the image to improve OCR results for Arabic text
        gray_img = ImageOps.invert(gray_img)

        # Enhance the contrast and sharpness of the image
        enhancer = ImageEnhance.Contrast(gray_img)
        gray_img = enhancer.enhance(2.0)
        gray_img = gray_img.filter(ImageFilter.SHARPEN)

        # Use Tesseract to do OCR on the preprocessed image
        text = pytesseract.image_to_string(gray_img, lang=lang)
        return text
    except Exception as e:
        return str(e)

# Example usage
image_path = 'download.jpg'  # Replace with the path to your image
text = ocr_image_to_text(image_path)
print(text)
```

```python
import cv2
import pytesseract
from PIL import Image
import numpy as np
import os


os.chdir(r'C:\Users\Games king\Desktop\OCR')
# Path to your Tesseract executable (if not in your PATH environment variable)
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

# Path to the Tesseract executable (if not in your PATH, specify the path directly)
# Uncomment and modify the following line if necessary
# pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

def preprocess_image(image_path):
    """
    Preprocess the image for better OCR results.

    Args:
    image_path (str): The file path to the image.

    Returns:
    PIL.Image: The preprocessed image.
    """
    # Read the image using OpenCV
    image = cv2.imread(image_path)
    
    # Convert to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Apply thresholding
    _, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
    
    # Remove noise
    kernel = np.ones((1, 1), np.uint8)
    opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel)
    
    # Resize the image
    resized_image = cv2.resize(opening, None, fx=1.5, fy=1.5, interpolation=cv2.INTER_LINEAR)
    
    # Convert back to PIL Image for pytesseract
    preprocessed_image = Image.fromarray(resized_image)
    
    return preprocessed_image

def ocr_image_to_text(image_path, lang='ara'):
    """
    Converts the content of an image to text using OCR.

    Args:
    image_path (str): The file path to the image.
    lang (str): The language to use for OCR (default is 'ara' for Arabic).

    Returns:
    str: The extracted text from the image.
    """
    try:
        # Preprocess the image
        preprocessed_image = preprocess_image(image_path)
        
        # Use Tesseract to do OCR on the image
        text = pytesseract.image_to_string(preprocessed_image, lang=lang)
        return text
    except Exception as e:
        return str(e)

# Example usage
image_path = 'khaled.jpg'  # Replace with the path to your image
text = ocr_image_to_text(image_path)
print(text)
```

```python
import cv2
import pytesseract
from PIL import Image
import numpy as np
from scipy.ndimage import interpolation as inter
import os


os.chdir(r'C:\Users\Games king\Desktop\OCR')
# Path to your Tesseract executable (if not in your PATH environment variable)
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

# Path to the Tesseract executable (if not in your PATH, specify the path directly)
# Uncomment and modify the following line if necessary
# pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

def preprocess_image(image_path):
    """
    Preprocess the image for better OCR results.

    Args:
    image_path (str): The file path to the image.

    Returns:
    PIL.Image: The preprocessed image.
    """
    # Read the image using OpenCV
    image = cv2.imread(image_path)
    
    # Convert to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Apply adaptive thresholding
    adaptive_thresh = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                            cv2.THRESH_BINARY_INV, 11, 2)
    
    # Remove noise using morphological operations
    kernel = np.ones((1, 1), np.uint8)
    morph = cv2.morphologyEx(adaptive_thresh, cv2.MORPH_CLOSE, kernel)
    
    # Resize the image
    resized_image = cv2.resize(morph, None, fx=1.5, fy=1.5, interpolation=cv2.INTER_LINEAR)
    
    # Correct skew
    coords = np.column_stack(np.where(resized_image > 0))
    angle = cv2.minAreaRect(coords)[-1]
    if angle < -45:
        angle = -(90 + angle)
    else:
        angle = -angle
    (h, w) = resized_image.shape[:2]
    center = (w // 2, h // 2)
    M = cv2.getRotationMatrix2D(center, angle, 1.0)
    rotated = cv2.warpAffine(resized_image, M, (w, h), flags=cv2.INTER_CUBIC, borderMode=cv2.BORDER_REPLICATE)
    
    # Sharpen the image
    sharpen_kernel = np.array([[-1, -1, -1], [-1, 9, -1], [-1, -1, -1]])
    sharpened = cv2.filter2D(rotated, -1, sharpen_kernel)
    
    # Convert back to PIL Image for pytesseract
    preprocessed_image = Image.fromarray(sharpened)
    
    return preprocessed_image

def ocr_image_to_text(image_path, lang='ara'):
    """
    Converts the content of an image to text using OCR.

    Args:
    image_path (str): The file path to the image.
    lang (str): The language to use for OCR (default is 'ara' for Arabic).

    Returns:
    str: The extracted text from the image.
    """
    try:
        # Preprocess the image
        preprocessed_image = preprocess_image(image_path)
        
        # Custom OCR configurations
        custom_config = r'--oem 3 --psm 6'
        
        # Use Tesseract to do OCR on the image
        text = pytesseract.image_to_string(preprocessed_image, lang=lang, config=custom_config)
        return text
    except Exception as e:
        return str(e)

# Example usage
image_path = 'khaled.jpg'  # Replace with the path to your image
text = ocr_image_to_text(image_path)
print(text)
```
